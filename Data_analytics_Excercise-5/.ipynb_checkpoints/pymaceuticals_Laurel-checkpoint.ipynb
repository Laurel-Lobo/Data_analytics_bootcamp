```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data
mouse_df = pd.read_csv(mouse_drug_data_to_load)
trial_df = pd.read_csv(clinical_trial_data_to_load)


# Combine the data into a single dataset
mouse_trial_df = pd.DataFrame.merge(mouse_df, trial_df, on = "Mouse ID")


# Display the data table for preview

mouse_trial_df.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint
# Convert to DataFrame
DrugTumorData = mouse_trial_df[["Drug", "Timepoint", "Tumor Volume (mm3)"]]

DrugTimeMeanTumor_df = DrugTumorData.groupby(["Drug", "Timepoint"], as_index = False)["Tumor Volume (mm3)"].mean().rename(columns={"Tumor Volume (mm3)" : "Mean Tumor Volume (mm3)"})


# Preview DataFrame

DrugTimeMeanTumor_df.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
# Convert to DataFrame

DrugTimeSteTumor_df = DrugTumorData.groupby(["Drug", "Timepoint"])["Tumor Volume (mm3)"].sem().reset_index().rename(columns={"Tumor Volume (mm3)": "Standard Error Tumor Volume"})

# Preview DataFrame

DrugTimeSteTumor_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames

DrugTimeMeanTumor_Clean_df2 = DrugTimeMeanTumor_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Mean Tumor Volume (mm3)')
DrugTimeSteTumor_Clean_df2 = DrugTimeSteTumor_df.pivot(index = 'Timepoint', columns = 'Drug', values = 'Standard Error Tumor Volume')

# Preview that Reformatting worked
DrugTimeMeanTumor_Clean_df2.head()
DrugTimeSteTumor_Clean_df2.head()
```

```python
DrugTimeMeanTumor_Clean_df2.reset_index(inplace=True)
DrugTimeSteTumor_Clean_df2.reset_index(inplace=True)
```

```python
# Generate the Plot (with Error Bars)
markers = ["o", "v", "1", "s", "p", "*", "+", "x", "D", "X"]

plt.figure(figsize=(20,10))

for idx, name in enumerate(DrugTimeMeanTumor_Clean_df2.columns[1:]):
    plt.errorbar(DrugTimeMeanTumor_Clean_df2["Timepoint"], DrugTimeMeanTumor_Clean_df2[name], yerr = DrugTimeSteTumor_Clean_df2[name], linestyle = '--', marker = markers[idx])

##Figure formatting
plt.xlabel("Time (Days)")
plt.ylabel("Mean Tumor Volume (mm3)")
plt.title("Tumor Response to Treatment")
plt.grid()
plt.legend(loc='best')

# Save the Figure
plt.savefig("Tumor_Response_To_Treatment.jpeg")

# Show the Figure
plt.show()
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint
## Convert to DataFrame
DrugSiteData = mouse_trial_df[["Drug", "Timepoint", "Metastatic Sites"]]

DrugTimeMeanSites_df = DrugSiteData.groupby(["Drug", "Timepoint"], as_index = False)["Metastatic Sites"].mean().rename(columns={"Metastatic Sites" : "Mean Metastatic Sites"})

# Preview DataFrame
DrugTimeMeanSites_df.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
# Convert to DataFrame
DrugTimeSitesErr_df = DrugSiteData.groupby(["Drug", "Timepoint"])["Metastatic Sites"].sem().reset_index().rename(columns={"Metastatic Sites" : "Standard Error Metastatic Sites"})

# Preview DataFrame
DrugTimeSitesErr_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
DrugTimeMeanSites_Clean_df = DrugTimeMeanSites_df.pivot(index = "Timepoint", columns = "Drug", values = "Mean Metastatic Sites")
DrugTimeSitesErr_Clean_df = DrugTimeSitesErr_df.pivot(index = "Timepoint", columns = "Drug", values = "Standard Error Metastatic Sites")
# Preview that Reformatting worked
DrugTimeMeanSites_Clean_df.head()
DrugTimeSitesErr_Clean_df.head()
```

```python
#resets index to allow graphing
DrugTimeMeanSites_Clean_df.reset_index(inplace=True)
DrugTimeSitesErr_Clean_df.reset_index(inplace=True)
```

```python
# Generate the Plot (with Error Bars)
markList = ["o", "v", "1", "s", "p", "*", "+", "x", "D", "X"]

plt.figure(figsize=(20,10))

for idx, name in enumerate(DrugTimeMeanSites_Clean_df.columns[1:]):
    plt.errorbar(DrugTimeMeanSites_Clean_df["Timepoint"], DrugTimeMeanSites_Clean_df[name], yerr = DrugTimeSitesErr_Clean_df[name], linestyle = '--', marker = markList[idx])

##Figure formatting
plt.title("Metastatic spread During Treatment")
plt.xlabel("Treatment Duration (Days)")
plt.ylabel("Mean Met. Sites")
plt.legend(loc='best')
plt.grid()

# Save the Figure
plt.savefig("Metastatic Spread During Treatment.jpeg")
# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)
# Convert to DataFrame
DrugMouseCountData = mouse_trial_df[["Drug", "Timepoint", "Mouse ID"]]

DrugMouseCount_df = DrugMouseCountData.groupby(["Drug", "Timepoint"], as_index = False)["Mouse ID"].count().rename(columns={"Mouse ID": "Mouse Count"})

# Preview DataFrame
DrugMouseCount_df.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
DrugMouseCount_Clean_df = DrugMouseCount_df.pivot(index = "Timepoint", columns = "Drug", values = "Mouse Count")

# Preview the Data Frame
DrugMouseCount_Clean_df.head()
```

```python
DrugMouseCount_Clean_df.reset_index(inplace=True)
```

```python
# Generate the Plot (Accounting for percentages)
markList = ["o", "v", "1", "s", "p", "*", "+", "x", "D", "X"]

plt.figure(figsize=(20,10))
  


for idx, name in enumerate(DrugMouseCount_Clean_df.columns[1:]):
    plt.errorbar(DrugMouseCount_Clean_df["Timepoint"],
                 [(x/(max(DrugMouseCount_Clean_df[name]))*100) for x in DrugMouseCount_Clean_df[name]],
                 label = name,
                 linestyle = '--',
                 marker = markList[idx])

    
##Figure formatting
plt.title("Survival During Treatment")
plt.xlabel("Time (Days)")
plt.ylabel("Survival Rate (%)")
plt.legend(loc='best')
plt.grid()

# Save the Figure
plt.savefig("Survival During Treatment.jpeg")

# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
#DrugTimeMeanTumor_Clean_df2["Capomulin"][9]-DrugTimeMeanTumor_Clean_df2["Capomulin"][9]

names = [name for name in DrugTimeMeanTumor_Clean_df2.columns[1:]]
value = [((DrugTimeMeanTumor_Clean_df2[name][9]-DrugTimeMeanTumor_Clean_df2[name][0])/DrugTimeMeanTumor_Clean_df2[name][0])*100 for name in DrugTimeMeanTumor_Clean_df2.columns[1:]]

DrugTumVolChange_df = pd.DataFrame({"Name": [name for name in DrugTimeMeanTumor_Clean_df2.columns[1:]],
                                    "Percent change (%)": [((DrugTimeMeanTumor_Clean_df2[name][9]-DrugTimeMeanTumor_Clean_df2[name][0])/DrugTimeMeanTumor_Clean_df2[name][0])*100 for name in DrugTimeMeanTumor_Clean_df2.columns[1:]]})
# Display the data to confirm
DrugTumVolChange_df
```

```python
plt.figure(figsize =(20,10))

# Splice the data between passing and failing drugs

good_drugs = DrugTumVolChange_df[DrugTumVolChange_df["Percent change (%)"] < 0]
bad_drugs = DrugTumVolChange_df[DrugTumVolChange_df["Percent change (%)"] > 0]

# Orient widths. Add labels, tick marks, etc. 
plt.bar(good_drugs["Name"], good_drugs["Percent change (%)"], color = 'g', width = 1.0)
plt.bar(bad_drugs["Name"], bad_drugs["Percent change (%)"], color = 'r', width = 1.0)

##Figure formatting
plt.title("Tumor Change Over 45 Day Treatment")
plt.xlabel("Drug Name")
plt.ylabel("% Tumor Volume Change")
plt.grid()


# Use functions to label the percentages of changes
def labelPercent (df):
    a = 0
    b = 2
    for x in df.iterrows():
        if x[1][1] < 0:
            plt.text(s = (f"{(round(x[1][1]))}%"), x = a, y = -4, color = "w", horizontalalignment = 'center', fontweight = 'bold')
            a += 1
        if x[1][1] > 0:
            plt.text(s = (f"{(round(x[1][1]))}%"), x = b, y = 4, color = "w", horizontalalignment = 'center', fontweight = 'bold')
            b += 1
    
# Call functions to implement the function calls

labelPercent(df = DrugTumVolChange_df)

# Save the Figure

plt.savefig("Tumor Change Over 45 Day Treatment.jpeg")

# Show the Figure
plt.show()
```
